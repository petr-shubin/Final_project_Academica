```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
```

```python
count_a, count_b = 1003, 1099
nobs_a, nobs_b = 100047501, 100001055

successes = np.array([count_a, count_b])
samples = np.array([nobs_a, nobs_b])

# Расчет конверсии в процентах
cr_a = count_a / nobs_a * 100
cr_b = count_b / nobs_b * 100
uplift = (cr_b - cr_a) / cr_a * 100

# Z-тест на пропорции
z_stat, p_value = proportions_ztest(successes, samples)

print(f"Конверсия А: {cr_a:.7f}%")
print(f"Конверсия B: {cr_b:.7f}%")
print(f"Относительный прирост: {uplift:.2f}%")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ")
else:
    print("Результат: НЕ ЗНАЧИМ")
```
